```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale
```

```python
wine= pd.read_csv("wine.csv")
wine
```

```python
wine1= wine.iloc[:,1:]
wine1
```

```python
wine.describe()
```

```python
wine.info()
```

```python
wine1.shape
```

# PCA building

```python
wine_ary=wine1.values
wine_ary
```

```python
# Normalizing 
wine_norm=scale(wine_ary)
wine_norm
```

# PCA Implementation

```python
pca = PCA()
pca_values = pca.fit_transform(wine_norm)
pca_values
```

```python
#amount of var
var = pca.explained_variance_ratio_
var
```

```python
#cumulative var
Var = np.cumsum(np.round(var,decimals= 4)*100)
Var
```

```python
wine_PCA=pd.concat([wine['Type'],pd.DataFrame(pca_values[:,0:3],columns=['PCA1','PCA2','PCA3'])],axis=1)
wine_PCA
```

```python
# Visualization of PCA
import seaborn as sns
fig=plt.figure(figsize=(16,15))
sns.scatterplot(data=wine_PCA);
```

```python
#Hierarchical Clustering
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import normalize
```

```python
plt.figure(figsize=(10,10))
dendrogram=sch.dendrogram(sch.linkage(wine_norm,'complete'))
```

```python
hclusters=AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='ward')
hclusters
```

```python
y=pd.DataFrame(hclusters.fit_predict(wine_norm),columns=['clusters id'])
y['clusters id'].value_counts()
```

```python
wine2=wine.copy()
wine2['clusters id']=hclusters.labels_
wine2
```

```python
#KMeans
from sklearn.cluster import KMeans
```

```python
wcss=[]
for i in range (1,6):
    kmeans=KMeans(n_clusters=i,random_state=2)
    kmeans.fit(wine_norm)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,6),wcss)
plt.title('Elbow Graph')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS');
```

```python
#Cluster algorithm using K=3
c_3=KMeans(3,random_state=30).fit(wine_norm)
c_3
```

```python
c_3.labels_
```

```python
wine3=wine.copy()
wine3['c_3_id']=c_3.labels_
wine3
```
